```python
# use make_blob to illustrate clusering
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import seaborn as sns

# globaly increase the font of matplotlib
plt.rcParams.update({'font.size': 18})
```

```python
X, y = make_blobs(n_samples=100, centers=3, n_features=2, random_state=42)
fig, ax = plt.subplots(2,1, figsize=(7,8), sharex=True)
plt.style.use("dark_background")
sns.scatterplot(x=X[:, 0], y=X[:, 1], palette='viridis', ax=ax[0])
ax[0].set_xlabel('X1')
ax[0].set_ylabel('X2')
ax[0].set_title('Original Data')
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
centroids = kmeans.cluster_centers_
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y_kmeans, palette='viridis', ax=ax[1])
# make an arrow to the one centroid
sns.scatterplot(x=centroids[:, 0], y=centroids[:, 1], color='red', s=100, ax=ax[1])
ax[1].set_title('Clustered Data using KMeans')
ax[1].set_xlabel('X1')
ax[1].set_ylabel('X2')
plt.tight_layout()
plt.savefig('kmeans_ilustration.pdf')
plt.show()
```
