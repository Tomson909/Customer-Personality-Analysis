### Drop x%

```python
# drop the 24 missing values
data.dropna(inplace=True)

display(data['Education'].value_counts())
display(data['Marital_Status'].value_counts())
display(data['Response'].value_counts())
display(data['Z_Revenue'].value_counts())
display(data['Z_CostContact'].value_counts())

# since the columns 'Z_Revenue' and 'Z_CostContact' have only one value, we can drop them
data.drop(columns=['Z_Revenue', 'Z_CostContact'], inplace=True)
# Dt_Customer to datetime, "21-08-2013"
data['Dt_Customer'] = pd.to_datetime(data['Dt_Customer'], format='%d-%m-%Y')
```

### Correlation Matrix of Features

```python
#labelencode
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
data_dummies = pd.get_dummies(data.drop(columns='Dt_Customer'), drop_first=True)
# correlation matrix
plt.figure(figsize=(20, 10))
mask = np.triu(data_dummies.corr(), k=1)
sns.heatmap(data_dummies.corr(), annot=True, fmt=".2f", cmap='icefire', mask=mask)
plt.title('Correlation Matrix')
plt.show()
```

- The Orders are correlated
- The Campaings also seem to be correlated. So that means that there are some groups of people, which maybe either fall for a campaign in general or not.
- Income is strongly correlated to purchases
- The Marital Status is slightly correlated with kinda everything.  

There is much more you can discuss about this correlation plot. 

### Categorical Columns

```python
cal_cols = ['Complain', 'Education', 'Marital_Status', 'Kidhome', 'Teenhome']
# Countplot for each categorical column
fig, ax = plt.subplots(1, len(cal_cols), figsize=(20, 5))
for i, col in enumerate(cal_cols):
    # countplot only works fors numerical data for some way. i will use value counts combines with barplot
    sns.barplot(x=data[col].value_counts().index, y=data[col].value_counts().values, ax=ax[i])
    # rotate
    ax[i].tick_params(axis='x', rotation=45)
    ax[i].set_title(col)

plt.show()
```

```python
# countplot of the date of births
plt.figure(figsize=(20, 5))
sns.barplot(x=data['Year_Birth'].value_counts().index, y=data['Year_Birth'].value_counts().values)
plt.tick_params(axis='x', rotation=45)
```

```python
# total mnt spent
data['Total_Mnt'] = data['MntWines'] + data['MntFruits'] + data['MntMeatProducts'] + data['MntFishProducts'] + data['MntSweetProducts'] + data['MntGoldProds']
# bar chart, y axis: total mnt, x axis: Response
plt.figure(figsize=(20, 5))
money_spend = data.groupby('Response')['Total_Mnt'].sum()
response_no_yes = data.groupby('Response')['Response'].count()
normalized_money_spend = money_spend / response_no_yes
sns.barplot(x=normalized_money_spend.index, y=normalized_money_spend.values, palette='viridis')
```

People who responded to the campaign also spend 100% more money. But that does not nececerraly mean, that the reason for that is the campaign. There could be another highly correlated feature actually causing this.
