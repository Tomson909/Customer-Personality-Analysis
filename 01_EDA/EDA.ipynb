### Correlation Matrix of Features

```python
#labelencode
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
data_dummies = pd.get_dummies(data.drop(columns='Dt_Customer'), drop_first=True)
# correlation matrix
plt.figure(figsize=(20, 10))
mask = np.triu(data_dummies.corr(), k=1)
sns.heatmap(data_dummies.corr(), annot=True, fmt=".2f", cmap='icefire', mask=mask)
plt.title('Correlation Matrix')
plt.show()
```

- The Orders are correlated
- The Campaings also seem to be correlated. So that means that there are some groups of people, which maybe either fall for a campaign in general or not.
- Income is strongly correlated to purchases
- The Marital Status is slightly correlated with kinda everything.
